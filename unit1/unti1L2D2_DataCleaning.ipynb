```python
# load data 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from tabulate import tabulate
import re
```

```python
with open('data/WELLCOME_APCspend2013_forThinkful.csv',encoding='UTF-16' ) as f:
    print(f)
```

```python
df=pd.read_csv('data/WELLCOME_APCspend2013_forThinkful.csv', encoding='Latin-1', error_bad_lines=False)
```

```python
df.head(989)
```

```python
df['Journal title']=df['Journal title'].astype('str') 
```

```python
df.info()
```

```python
cleanFun=lambda row: ' '.join(row['Journal title'].split()).upper().strip()
df['nt'] = df.apply(cleanFun , axis=1)
```

```python
print(df['nt']  ) 
```

```python
df.head(11)
```

```python
df['nt'] = df['nt'].replace(['PLOSONE', 'PLOS ONE', 'PLOS 1', 'PLOS'], 'PLOS ONE')
#print(df[df['nt'].str.contains('PLOS')]['nt'])
```

```python
df['nt'] = df['nt'].replace(['AMERICAN JOURNAL OF MEDICAL GENETICS PART A'], 'AMERICAN JOURNAL OF MEDICAL GENETICS') 
df['nt'] = df['nt'].replace(['AMERICAN JOURNAL OF MEDICAL GENETICS PART A'], 'AMERICAN JOURNAL OF MEDICAL GENETICS') 
df['nt'] = df['nt'].replace(['ANTIMICROBIAL AGENTS AND CHEMOTHERAPY', 'ANTIMICROBIAL AGFENTS AND CHEMOTHERAPY'], 'ANTIMICROBIAL AGENTS & CHEMOTHERAPY') 
df['nt'] = df['nt'].replace(['ANGEWANDE CHEMIE', 'ANGEWANDTE CHEMIE INTERNATIONAL EDITION','ANGEW CHEMS INT ED' ], 'ANGEWANDTE CHEMIE') 
df['nt'] = df['nt'].replace(['BEHAVIOUR RESEARCH AND THERAPY'], 'BEHAVIOR RESEARCH & THERAPY')

df.head(55)
```

```python
jp=df.groupby('nt').count().sort_values('Article title',ascending=False)[0:5]
```

#### calculate the mean, median, and standard deviation of the open-access cost per article for each journal . 
   

```python
df=df.rename(columns={'COST (£) charged to Wellcome (inc VAT when charged)': 'GB_price'})
df.head(5)
```

```python
#need find first or one need string not list 
p = re.compile('\d+\.?\d+') 
rateFn=lambda row: float(p.findall(row['GB_price'])[0])
df['price'] = df.apply(rateFn , axis=1)
```

```python
df.drop(df[df.price > 99999].index, inplace=True)
```

```python
jsts=df.groupby('nt').agg({'price':['mean','median','std']})
```

```python
print(tabulate(jsts[0:10], tablefmt="github", headers="keys"))
```
